```python


def plot_res(telescope,file):
    f=open(file)
    #f=open('/home/tasha/work/1012-4235/timing/residuals/out.res')
    year=[]
    freq=[]
    mjd=[]
    pre_res_ph=[]
    pre_res_us=[]
    post_res_us=[]
    res_er_ph=[]
    res_er_us=[]
    orb_ph=[]
    for line in f:
        line=line.strip()
        columns = line.split()
        #Parkes data:
        if telescope=='mock1':
            if float(columns[1]) < 55817 and float(columns[1]) > 55407:
                #if float(columns[2]) > 1000:
                year.append(float(columns[0]))
                mjd.append(float(columns[1]))
                freq.append(float(columns[2]))
                pre_res_ph.append(float(columns[3]))
                pre_res_us.append(float(columns[4])) 
                post_res_us.append(float(columns[5]))  
                res_er_ph.append(float(columns[6]))  
                res_er_us.append(float(columns[7]))
                orb_ph.append(float(columns[8]))
        elif telescope=='mock2':
            if float(columns[1]) < 56219 and float(columns[1]) > 55888:
                #if float(columns[2]) < 1000:
                
                year.append(float(columns[0]))
                mjd.append(float(columns[1]))
                freq.append(float(columns[2]))
                pre_res_ph.append(float(columns[3]))
                pre_res_us.append(float(columns[4])) 
                post_res_us.append(float(columns[5]))  
                res_er_ph.append(float(columns[6]))  
                res_er_us.append(float(columns[7]))
                orb_ph.append(float(columns[8]))
        elif telescope=='puppi':
            if float(columns[1]) < 59054 and float(columns[1]) > 58880:
                #if float(columns[7]) < 80:
                #if float(columns[2]) > 1100:
                    year.append(float(columns[0]))
                    mjd.append(float(columns[1]))
                    freq.append(float(columns[2]))
                    pre_res_ph.append(float(columns[3]))
                    pre_res_us.append(float(columns[4])) 
                    post_res_us.append(float(columns[5]))  
                    res_er_ph.append(float(columns[6]))  
                    res_er_us.append(float(columns[7]))
                    orb_ph.append(float(columns[8]))
        elif telescope=='fast':
            if float(columns[1]) < 59555 and float(columns[1]) > 59506:
                year.append(float(columns[0]))
                mjd.append(float(columns[1]))
                freq.append(float(columns[2]))
                pre_res_ph.append(float(columns[3]))
                pre_res_us.append(float(columns[4])) 
                post_res_us.append(float(columns[5]))  
                res_er_ph.append(float(columns[6]))  
                res_er_us.append(float(columns[7]))
                orb_ph.append(float(columns[8]))
        
        elif telescope=='freq-divide-mid':
            if float(columns[2]) < 1.3 and float(columns[2]) > 1.6:
                year.append(float(columns[0]))
                mjd.append(float(columns[1]))
                freq.append(float(columns[2]))
                pre_res_ph.append(float(columns[3]))
                pre_res_us.append(float(columns[4])) 
                post_res_us.append(float(columns[5]))  
                res_er_ph.append(float(columns[6]))  
                res_er_us.append(float(columns[7]))
                orb_ph.append(float(columns[8]))
        elif telescope=='freq-divide-low':
            if float(columns[2]) < 1.3:
                year.append(float(columns[0]))
                mjd.append(float(columns[1]))
                freq.append(float(columns[2]))
                pre_res_ph.append(float(columns[3]))
                pre_res_us.append(float(columns[4])) 
                post_res_us.append(float(columns[5]))  
                res_er_ph.append(float(columns[6]))  
                res_er_us.append(float(columns[7]))
                orb_ph.append(float(columns[8]))
        elif telescope=='freq-divide-high':
            if float(columns[2]) > 1.6:
                year.append(float(columns[0]))
                mjd.append(float(columns[1]))
                freq.append(float(columns[2]))
                pre_res_ph.append(float(columns[3]))
                pre_res_us.append(float(columns[4])) 
                post_res_us.append(float(columns[5]))  
                res_er_ph.append(float(columns[6]))  
                res_er_us.append(float(columns[7]))
                orb_ph.append(float(columns[8]))
        elif telescope=='all':
            
            year.append(float(columns[0]))
            mjd.append(float(columns[1]))
            freq.append(float(columns[2]))
            pre_res_ph.append(float(columns[3]))
            pre_res_us.append(float(columns[4])) 
            post_res_us.append(float(columns[5]))  
            res_er_ph.append(float(columns[6]))  
            res_er_us.append(float(columns[7]))
            orb_ph.append(float(columns[8]))
    return mjd,year,orb_ph,post_res_us,res_er_us
```

```python
from matplotlib.ticker import MaxNLocator


f='/home/tasha/work/1952+2630/final/out.res'

#plt.figure(figsize=(10,5))
#plt.rc('axes', labelsize=15)
#plt.rc('xtick', labelsize=15)
#plt.rc('ytick', labelsize=15)
#plt.xlabel('Year')
#plt.ylabel('Residual ($\mu$s)')
fig, ax1 = plt.subplots(figsize=(15,6),tight_layout=True) 
#fig = plt.figure()
plt.rc('xtick', labelsize=25)
plt.rc('ytick', labelsize=25)

mjd,year,orb_ph,post_res_us,res_er_us= plot_res('mock1',f)
ax1.errorbar(mjd,post_res_us,yerr=res_er_us,color='goldenrod',ms=8,fmt='.',label='Mock-1')
ax1.legend(fontsize=12)


mjd,year,orb_ph,post_res_us,res_er_us = plot_res('mock2',f)
ax1.errorbar(mjd,post_res_us,yerr=res_er_us,color='lightseagreen',ms=8,fmt='.',label='Mock-2')
ax1.legend(fontsize=12)

mjd,year,orb_ph,post_res_us,res_er_us = plot_res('puppi',f)
print(len(mjd))
ax1.errorbar(mjd,post_res_us,yerr=res_er_us,color='rosybrown',ms=8,fmt='.',label='Puppi')
ax1.legend(fontsize=12)

mjd,year,orb_ph,post_res_us,res_er_us = plot_res('fast',f)
ax1.errorbar(mjd,post_res_us,yerr=res_er_us,color='gray',ms=8,fmt='.',label='ROACH')
ax1.legend(fontsize=12)

period=0.0207323600913747
mjd,year,orb_ph,post_res_us,res_er_us = plot_res('all',f)

post_res_phase=[]
res_er_phase=[]
for r in post_res_us:
    post_res_phase.append(r*10**(-6)/period)

for er in res_er_us:
    res_er_phase.append(er*10**(-6)/period)

ax1.errorbar(mjd,post_res_us,yerr=res_er_us,color='none',fmt='.')
ax2=ax1.twiny()
plt.rc('xtick', labelsize=25)
plt.rc('ytick', labelsize=25)
ax2.errorbar(year,post_res_us,yerr=res_er_us,color='none',fmt='.')
ax2.set_xticks([2010,2012,2014,2016,2018,2020,2022])
ax3=ax1.twinx()
plt.rc('xtick', labelsize=25)
plt.rc('ytick', labelsize=25)
ax3.errorbar(mjd,post_res_phase,yerr=res_er_phase,color='none',fmt='.')


#ax1.set_ylim(-200,200)
#ax1.set_xlim(58500,59500)
ax1.set_xlabel('MJD',fontsize=25)
ax1.set_ylabel('Residual ($\mu$s)',fontsize=25)
ax2.set_xlabel('Year',fontsize=25)
ax3.set_ylabel('Residual (phase)',fontsize=25)

fig.savefig("residual-plot-1952-mjd.pdf", format='pdf')
```
