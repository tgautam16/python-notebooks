```python
import pylab
import matplotlib.pyplot as plt
import psrchive
import numpy as np
```

```python
#! /usr/bin/env python

def plot_int_profile(filename):
    filename=str(filename)
    arch = psrchive.Archive_load(filename)
    arch.dedisperse()
    arch.remove_baseline()
    data = arch.get_data()
    I=data[:,:,:,:].mean(0).mean(0).mean(0)
    shp=data.shape
    nbin=shp[3]
    max_int=np.argmax(I)
    if max_int<int(nbin/2):
        #print((nbin/2)-max_int)
        I=np.roll(I,int((nbin/2)-max_int))
        #max_int=np.argmax(I_new)
    else:
        #print(max_int)
        #print((max_int-(nbin/2)))
        I=np.roll(I,-int(max_int-(nbin/2)))
    
    fig = plt.figure(figsize=(5,5))
    plt.yticks([])
    plt.rc('font', size=15)          # controls default text sizes
    plt.rc('axes', titlesize=15)     # fontsize of the axes title
    plt.rc('axes', labelsize=15) 
    plt.tight_layout()

    plt.plot(np.arange(0,1,1./nbin),I,linewidth=2.0,c='k')
    plt.xlabel('Pulse phase')
    plt.ylabel('Flux density (relative units)')
    srcname=arch.get_source()
    srcname=srcname.replace("_","J")
    plt.title(srcname)
    plt.savefig(str(arch.get_source())+'.pdf',bbox_inches='tight')  
    
def plot_int_profile_nomove(filename,bin_move,name):
    filename=str(filename)
    arch = psrchive.Archive_load(filename)
    arch.dedisperse()
    arch.remove_baseline()
    data = arch.get_data()
    I=data[:,:,:,:].mean(0).mean(0).mean(0)
    shp=data.shape
    nbin=shp[3]
    max_int=np.argmax(I)
    I=np.roll(I,bin_move)
    fig = plt.figure(figsize=(5,5))
    
    plt.yticks([])
    plt.rc('font', size=15)          # controls default text sizes
    plt.rc('axes', titlesize=15)     # fontsize of the axes title
    plt.rc('axes', labelsize=15) 
    plt.tight_layout()
    plt.plot(np.arange(0,1,1./nbin),I,label=arch.get_centre_frequency(),linewidth=2.0,c='k')
    plt.xlabel('Pulse phase')
    plt.ylabel('Flux density (relative units)')
    srcname=arch.get_source()
    srcname=srcname.replace("_","J")
    plt.title(srcname)
    plt.legend()
    plt.savefig(str(arch.get_source())+'_'+name+'.pdf',bbox_inches='tight')  
```

```python
with open('archive_list.txt') as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]
    #plot_int_profile(lines)
    for line in lines:
        plot_int_profile(line)


#sub,pol,chan,bin
```
